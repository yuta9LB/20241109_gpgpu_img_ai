## GoogleColaboratoryについて

GoogleColabratory(以下Colab)ではPythonをインストールすることなくプログラミングの実行環境をWeb上で実現することができます。

Colabでは、ブラウザ上で Python を記述、実行できます。以下の機能を使用できます。
- 環境構築が不要
- GPU に料金なしでアクセス
- 簡単に共有

## 機能制限について

・RAM：12GB

・ディスク：CPU/TPC:最大107GB、GPU:最大68GB

・90分ルール : 何も操作せずに90分経つとリセット

・12時間ルール : インスタンスが起動してから12時間経つとリセット

・GPUの使用制限  : GPUを使いすぎるとリセット（上限未公開）

# コードを自動生成してみる

ColabではAIを用いてコーディングを委託することもできます。サンプルを以下に示します。

<img src="img/img4.png" width = 30%>

「生成」をクリックし、「sin波のグラフを描画してください。」と入力します。

<img src="img/img5.png" width = 30%>


```python
# prompt: sin波のグラフを描画してください。

import numpy as np
import matplotlib.pyplot as plt

# xの値を生成
x = np.linspace(0, 2 * np.pi, 100)

# sin波の計算
y = np.sin(x)

# グラフを描画
plt.plot(x, y)

# グラフのタイトルと軸ラベルを設定
plt.title('Sin Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# グリッドを表示
plt.grid(True)

# グラフを表示
plt.show()
```

以上のように、AIがsin波のグラフを表示するプログラムを作成してくれます。

このようにテンプレートとしてAIにプログラムを出力させ、細かい部分を修正したり調べたりしながら完成させて効率よくプログラムを作成する方法もあります。
